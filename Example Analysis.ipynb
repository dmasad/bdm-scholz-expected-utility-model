# BDM-Scholz Expected Utility Model

Quick example of how to use Jeremy McKibben-Sanders's [replication](https://github.com/jmckib/bdm-scholz-expected-utility-model) of the [Scholz et al. implementation](http://jtp.sagepub.com/content/23/4/510) of Bueno de Mesquita's Expected Utility Model.

```python
import random

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from bdm_scholz_model import BDMScholzModel
```

# Run Once

```python
model = BDMScholzModel.from_csv_path("ExampleActors.csv", verbose=False)
```

```python
median_positions = []
mean_positions = []
```

```python
for _ in range(10):
    median_positions.append(model.median_position())
    mean_positions.append(model.mean_position())
    model.run_model(1)
```

```python
fig, ax = plt.subplots(figsize=(8, 4))
l1, = ax.plot(median_positions, lw=2)
l2, = ax.plot(mean_positions, lw=2)
ax.legend([l1, l2], ["Medians", "Means"], loc=(1.01, 0))
ax.grid(True)
_ = ax.set_ylim(5, 10.5) 
```

# Sensitivity Analysis

Run the model $N$ times; each time, perturb the saliences by $\epsilon \sim Normal(0, \sigma)$ and see what the distribution of median positions looks like. 

```python
N = 10
sigma = 0.05 
```

### Baseline run

```python
model = BDMScholzModel.from_csv_path("ExampleActors.csv", verbose=False)
baseline_medians = []
for _ in range(10):
    baseline_medians.append(model.median_position())
    model.run_model(1)
baseline_medians.append(model.median_position())
```

### Perturbed Runs

```python
def perturb_s(s, sigma):
    '''Perturb s ~ Normal(0, sigma), st s still in [0, 1].
    
    Returns:
        New value of s
    '''
    new_s = s + random.normalvariate(0, sigma)
    new_s = max(0, min(new_s, 1))
    return new_s
```

```python
all_medians = []
for i in range(N):
    medians = []
    model = BDMScholzModel.from_csv_path("ExampleActors.csv", verbose=False)
    # Perturb saleinces
    for a in model.actors:
        a.s = perturb_s(a.s, sigma)
    # Run
    for _ in range(10):
        medians.append(model.median_position())
        model.run_model(1)
    medians.append(model.median_position())
    all_medians.append(medians)
    print("Run {} done.".format(i))
```

```python
fig, ax = plt.subplots(figsize=(8, 4))

for medians in all_medians:
    l1, = ax.plot(medians, color='k', alpha=0.2)

l2, = ax.plot(baseline_medians, color='blue')
ax.legend([l2, l1], ["Original", "Perturbed"], loc=(1.01, 0))
ax.grid(True)
ax.set_ylim(5, 10.5)
```

```python
final_medians = [medians[-1] for medians in all_medians]

fig, ax = plt.subplots(figsize=(4, 4))
hist = ax.hist(final_medians, color='grey')
ax.vlines(baseline_medians[-1], 0, N, color='blue')
ax.set_ylim(0, N)
ax.set_xlim(5, 10)
```
